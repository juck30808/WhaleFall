```python
import networkx as nx
import plotly.express as px
import matplotlib.pyplot as plt
data = px.data.gapminder()
data.head()
```

```python
data['nation'] = 'Country'
data = nx.from_pandas_edgelist(data, source='nation', target='country', edge_attr=True,)
```

```python
import warnings
#warnings.filterwarnings('ignore')

plt.rcParams['figure.figsize']  = (22, 20)
plt.style.use('fivethirtyeight')
plt.title('Countries in the World', fontsize = 30)
plt.grid()
plt.axis('off')

pos = nx.spring_layout(data)
nx.draw_networkx_nodes(data, pos, node_size = 1200, node_color = 'pink') # drawing nodes
nx.draw_networkx_labels(data, pos, font_size = 20, font_family = 'sans-serif') # labels
nx.draw_networkx_edges(data, pos, width = 6, alpha = 0.1, edge_color = 'brown') # drawing edges

plt.show()
```
