## 3+ema+channels+screening+(c)

```python
import yfinance as yf
import datetime as dt
import pandas as pd 
import numpy as np

universe = 'Nasdaq100'
#'Nasdaq100','sp500'
```

```python
def calculate_ema(prices, days, smoothing=2):
    ema = [sum(prices[:days]) / days]
    for price in prices[days:]:
        ema.append((price * (smoothing / (1 + days))) + ema[-1] * (1 - (smoothing / (1 + days))))
    return ema
```

```python
prev = 5
```

```python
tickers = []
ema = []
if universe == 'Nasdaq100':
    df = pd.read_excel(r'input\Nasdaq100.xlsx',header=None)
    df.columns = ['Nasdaq']
    tickers = df['Nasdaq'].values.tolist()
    
elif universe == 'sp500':
    df = pd.read_excel(r'input\sp500.xlsx',header=None)
    df.columns = ['sp']
    tickers = df['sp'].values.tolist()
    
start = dt.datetime.today()-dt.timedelta(600)
end = dt.datetime.today()
ohlc_intraday = {}

for ticker in tickers:
    ohlc_intraday[ticker] = yf.download(ticker,start,end) 
        
for i in range(0,prev):
    add_ema = []
    for ticker in tickers:
        noskip = True
        if len(ohlc_intraday[ticker]["Close"])>200 + i:
            close = ohlc_intraday[ticker]["Close"][-1-i]
            low_t_1 = ohlc_intraday[ticker]["Low"][-2-i]
            
                
            ma50close = calculate_ema(ohlc_intraday[ticker]["Close"],50)[-2-i]
            ma100close = calculate_ema(ohlc_intraday[ticker]["Close"],100)[-2-i]
            ma150close = calculate_ema(ohlc_intraday[ticker]["Close"],150)[-2-i]       
            ma50close_t_1 = calculate_ema(ohlc_intraday[ticker]["Close"],50)[-3-i]                
        else:
            noskip = False

        if noskip and (close >= ma50close)and(ma50close > ma100close)and(ma100close > ma150close)and(low_t_1 <ma50close_t_1):
            add_ema.append(ticker)

    ema.append(add_ema) 
    
```
