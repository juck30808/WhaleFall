```python
import pandas as pd 
import numpy as np
sales_arrays = [['sale1', 'sale1', 'sale2', 'sale2', 'sale3', 'sale3', 'sale4', 'sale4'],
          ['city1', 'city2', 'city1', 'city2', 'city1', 'city2', 'city1', 'city2']]
sales_tuples = list(zip(*sales_arrays))
sales_index = pd.MultiIndex.from_tuples(sales_tuples, names=['sale', 'city'])

print("\nConstruct a Dataframe using the said MultiIndex levels: ")
df = pd.DataFrame(np.random.randn(8, 5), index=sales_index)
print(df)

print("\nExtract a single row from the given dataframe:")
print(df.loc[('sale2', 'city2')])

print("\nExtract number of rows from the given dataframe:")
print(df.loc['sale1'])

print("\nExtract a single value from the given dataframe:")
print(df.loc[('sale1', 'city2'), 1])
```

```python
import pandas as pd
rectangles = [
    { 'height': 40, 'width': 10 },
    { 'height': 20, 'width': 9 },
    { 'height': 3.4, 'width': 4 }
]

rectangles_df = pd.DataFrame(rectangles)

def calculate_area(row):
    return row['height'] * row['width']
def calculate_perimeter(row):
    return 2*(row['height']+row['width'])

rectangles_df['Area'] = rectangles_df.apply(calculate_area, axis=1)
rectangles_df['Perimeter']=rectangles_df.apply(calculate_perimeter, axis=1)
rectangles_df
```

```python
import pandas as pd
url = 'https://tinyurl.com/2hzvbjju'
drinks = pd.read_csv(url)

def calculate(drinks):
    return drinks['beer_servings'] + drinks['spirit_servings'] + drinks['wine_servings']

drinks['total_servings']=drinks.loc[:, 'beer_servings':'wine_servings'].apply(calculate, axis=1)

drinks['beer_sales']=drinks['beer_servings'].apply(lambda x: x*2)
drinks['spirit_sales']=drinks['spirit_servings'].apply(lambda x: x*4)
drinks['wine_sales']=drinks['wine_servings'].apply(lambda x: x*6)
drinks
```

```python
import pandas as pd
import numpy as np
matrix = [(22, 34, 23),
         (33, 31, 11),
         (44, 16, 21),
         (55, 32, 22),
         (66, 33, 27),
         (77, 35, 11)
         ]
# Create a DataFrame object
df = pd.DataFrame(matrix, columns=list('xyz') , index=list('abcdef'))

print(f'original data\n{df}')

# 1. square all the values in column ‘z’
df['z'] = df['z'].apply(np.square)

# 2. multipy all the values in row ‘c’ by 10
df.loc['c'] = df.loc['c'].apply(lambda x: x*10)

# 3. Add 10 in all the values in column ‘y’ & ‘x’
df = df.apply(lambda x: x+10 if x.name == 'y' or x.name == 'x' else x)

print(f'\ndataframe after applying modifications\n{df}')
```
