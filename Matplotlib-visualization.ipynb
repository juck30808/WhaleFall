```python
import matplotlib.pyplot as plt
from matplotlib import style

style.use("bmh")
#style.use("classic")
#style.use("dark_background")


x= [2,4,8]
y= [4,8,24]

x2= [3,6,9]
y2= [18,15,38]

plt.plot(x,y)
plt.plot(x2,y2)

plt.title("test")
plt.xlabel('test x')
plt.ylabel('test y')
plt.show()       #Show at same time
```

```python
import matplotlib.pyplot as plt
from matplotlib import style

style.use("bmh")

x= [2,4,8]
y= [4,8,24]

x2= [3,6,9]
y2= [18,15,38]

x3= [5,6,9]
y3= [15,15,38]

plt.bar(x,y)
plt.bar(x2,y2)
plt.bar(x3,y3)

plt.title("test")
plt.xlabel('test x')
plt.ylabel('test y')
plt.show()       #Show at same time
```

```python
import matplotlib.pyplot as plt

food = ['pizza','ice cream','burgers']
sales = [10,20,30]
color = ['red','blue','yellow']

plt.title("test")

plt.pie(sales,labels=food,colors=color)  
#plt.hist(x,y,histtype="step") 
plt.show()       #Show at same time
```
