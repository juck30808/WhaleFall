## 直線圖 

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

x= [2,4,8]
y= [4,8,24]

x2= [3,6,9]
y2= [18,15,38]

plt.plot(x,y)
plt.plot(x2,y2)
plt.show()       #Show at same time
```

## 長條圖 Bar chart

```python
import matplotlib.pyplot as plt
from matplotlib import style

style.use("bmh")

x= [2,4,8]
y= [4,8,24]

x2= [3,6,9]
y2= [18,15,38]

x3= [5,6,9]
y3= [15,15,38]

plt.bar(x,y)
plt.bar(x2,y2)
plt.bar(x3,y3)

plt.title("test")
plt.xlabel('test x')
plt.ylabel('test y')
plt.show()       #Show at same time
```

```python
import matplotlib.pyplot as plt


x= [2,4,8,3,6,9,3,6,9,15,15,4,8,24,8,3,6,9]
y= [3,4,8,9,15,20,25,30]

plt.title("test")
plt.xlabel('test x')
plt.ylabel('test y')

plt.hist(x,y,histtype="bar")  
#plt.hist(x,y,histtype="step") 
plt.show()       #Show at same time
```

## Pie Chart

```python
import matplotlib.pyplot as plt

food = ['pizza','ice cream','burgers']
sales = [10,20,30]
color = ['red','blue','yellow']

plt.title("test")

plt.pie(sales,labels=food,colors=color)  
#plt.hist(x,y,histtype="step") 
plt.show()       #Show at same time
```

```python
labels='India','US','UK','Germany'
sizes=[38,32,20,10]
explode=(0.1,0.2,0,0)

plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',shadow=True, startangle=90)
plt.axis('equal')
plt.show()
```

## scatter 散佈圖

```python
import matplotlib.pyplot as plt

x= [2,4,8,3,6,9]
y= [3,4,8,9,1,9]

plt.title("test")
plt.xlabel('test x')
plt.ylabel('test y')
plt.scatter(x,y)  
plt.show()       #Show at same time
```

```python
x=[2,5,7,6]
y=[12,34,44,78]
plt.scatter(x,y)
plt.show()
```

```python
N=50
x=np.random.rand(N)
y=np.random.rand(N)
colors=np.random.rand(N)
area=np.pi * (15 * np.random.rand(N))**2

plt.scatter(x,y, s=area, c=colors, alpha=0.5)
plt.show()
```

```python
N=50
x=np.array([1,2,4,5])
y=np.array([23,12,45,32])

plt.plot(x,y,'b--',linewidth=4)
plt.show()
```

```python
mu=100
sigma=15
x=mu+sigma*np.random.randn(10000)

n,bins,patches=plt.hist(x,50,density=True, facecolor='blue',alpha=0.80)

plt.xlabel('Smart graphs')
plt.ylabel('Probabilities')
plt.axis([40,160,0,0.03])
plt.grid(True)

plt.show()
```
