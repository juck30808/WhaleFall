## 默認瀏覽器顯示 url
webbrowser.open(url, new=0, autoraise=True)
如果 new 為 0，則在同一瀏覽器窗口中打開 url。如果 new 為 1，則盡可能打開新的瀏覽器窗口。如果 new 為 2，則盡可能打開新的瀏覽器頁面（“標籤”）。如果 autoraise 為 “True”，則會盡可能置前窗口（請注意，在許多窗口管理器下，無論此變量的設置如何，都會置前窗口）。

```python
import requests, sys, webbrowser, bs4

print('Googling...') 
# res = requests.get('http://google.com/search?q=' + ' '.join(sys.argv[1:]))
# res.raise_for_status()

name = "google"
res = requests.get('http://google.com/search?q=' + name)

# Retrieve top search result links.
soup = bs4.BeautifulSoup(res.text)
soup
```

```python
# Open a browser tab for each result.
linkElems = soup.select('a')  #.r a
linkElems
```

```python
numOpen = min(5, len(linkElems))
numOpen
```

```python
for i in range(2,numOpen):
    webbrowser.open('http://google.com' + linkElems[i].get('href'))
```
