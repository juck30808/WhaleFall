## Seaborn

```python
df = pd.read_csv("input/iris.csv")
df
```

```python
df2 = df[df["class"].isin(["Iris-setosa"])] # 篩選一下資料好做舉例
df2
```

```python
#pip install seaborn
```

```python
# 盒狀圖

import seaborn as sns
sns.boxplot(x=df2["class"], y=df2["Petal Width"])
```

```python
# 提琴圖
sns.violinplot(x=df2["Petal Length"], y=df2["Petal Width"])
```

```python
# 分佈圖
sns.regplot(x=df2["Petal Length"], y=df2["Petal Width"],ci=10)
```

```python
import matplotlib.pyplot as plt
from pandas.plotting import parallel_coordinates
iris = sns.load_dataset('iris') 
parallel_coordinates(iris, 'species')
plt.show()
```
