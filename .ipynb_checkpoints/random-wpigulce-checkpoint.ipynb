```python
import sys
import random

# "To jest mój pierwszy program w Pythonie"

#print() - wyświetla dane na ekran

print("To jest mój pierwszy program w Pythonie")



# Zmienna to miejsce do przechowywania wartości
# Nazwa zmiennej jest jak etykieta dla tej wartości

nick = "Trener Pythona"
print(nick)

# Nazwa zmiennej może zawierać litery i cyfry
# Nie może zaczynać się od liczby
# Istnieje 5 typów danych: Numbers(liczby), Strings(łańcuchy znaków), List(listy), Tuples(krotki), Dictionary(słownik)
# Możesz przechowywać dowolne z nich w tej samej zmiennej

liczba = 15
print(liczba)

#Operatory arytmetyczne +,-,*,/,%,**,//
#dodawanie -> +
#odejmowanie -> -
#mnożenie -> *
#dzielenie -> /
#reszta z dzielenia, modulo -> %

# Obliczenia wykładnicze
print("2 + 2 =", 2 + 2)
print("2 - 2 =", 2 - 2)
print("2 * 2 =", 2 * 2)
print("2 / 2 =", 2 / 2)
print("20 ** 2 =", 20 ** 2)
print("100 // 2 =", 100 // 2)


# Kolejność działań * i / jest wykonywana przed + i -
print("5 + 5 - 5 * 5 =", 5 + 5 - 5 * 5)
print("(5 + 5 - 5) * 5 =", (5 + 5 - 5) * 5)

#Jeśli musisz użyć znaku "cudzysłów", użyj znaku '\' w celu jego wyświetlenia

motto = "motto: \"Jedynym sposobem nauczenia się nowego języka programowania jest pisanie w nim programów.\""

print(motto)

#Wielowierszowe motto
wielowierszowe_motto = '''Jedynym sposobem nauczenia się nowego języka programowania 
jest pisanie w nim programów.'''

print(wielowierszowe_motto)

#Aby osadzić stringa - łańcuch znaków na wyjściu w Consoli użyj %s
print("%s %s %s" % ('lubie to motto', motto, wielowierszowe_motto))

#Aby nie printować nowych linii użyj end = ""
print("lubie", end="   ")
print("pythona")

#Możesz wyświetlić ciąg wiele razy za pomocą *
print('\n' * 5)

#LISTY
#lista umożliwia utworzenie listy wartości i manipulowanie nimi
#Każda wartość ma indeks przy czym pierwszy zaczyna się od 0!

lista_zakupów = ['Jajka','Ser','Pomidorki','Chilli']
print('Najważniejszy składnik jajecznicy to  ->  ', lista_zakupów[0])
print('Dodatkowy składnik jajecznicy to ->', lista_zakupów[2])

#Możesz zmienić wartość przechowywaną w polu listy:
lista_zakupów[0] = "Warzywa na patelenię"
print(lista_zakupów)

#Możesz zmienić wartości w taki sposób, aby uzyskać podzbiór listy

print(lista_zakupów[1:3])

#Możesz umieścić dowolny typ danych na liście łącznie z listą
inne_czynności = ['Sprzątanie domu', 'oglądanie telewizji', 'nauka programowania']
do_zrobienia = [inne_czynności, lista_zakupów]

print(do_zrobienia)

#Pobierz drugą pozycję na drugiej liście

print(do_zrobienia[1][1])

#Dodaj wartości za pomocą "append"
lista_zakupów.append('cytrynka')
print(do_zrobienia)

#Wstaw element pod podanym indeksem
lista_zakupów.insert(3,"ketchup")
print(lista_zakupów)

#Usuń propozycję z listy
lista_zakupów.remove("ketchup")
print(lista_zakupów)

#Sortuj pozycję na liście
lista_zakupów.sort()
print(lista_zakupów)

#Odwróć sortowanie elementów na liście
lista_zakupów.reverse()
print(lista_zakupów)

#Usuń element o określonym indeksie
del lista_zakupów[4]
print(lista_zakupów)

#Możemy łączyć listy za pomocą +
do_zrobienia = inne_czynności + lista_zakupów
print(do_zrobienia)

#Pozyskaj długość listy
print(len(do_zrobienia))

#Pozyskaj maksymalną pozycję na liście
print(max(do_zrobienia))

#Pozyskaj minimalną pozycję na liście
print(min(do_zrobienia))

# TUPLES / Krotki Wartości w krotce nie mogą zmieniać się jak listy
# Dla kontrastu elementy listy są numerowane od zera
# Listy są typem modyfikowalnym w związku z czym można dowolnie zmieniać ich elementy.
# Krotka jest ogólnie rzecz biorąc listą, której nie da się zmodyfikować.

pi_krotka =(3,1,4,1,5,9)

#Konwertuj krotkę na listę
nowa_krotka = list(pi_krotka)

#Krotki mają również len (tuple), min (tuple) i max (tuple)
print(len(pi_krotka))

#DICTIONARY / MAP
# Składa się z wartości z unikalnym kluczem dla każdej wartości
# Podobny do list
super_bohaterowie = {'Batman':'Supermen','Iron Men': 'Robin','Wizard': 'Deadpool'}

print(super_bohaterowie)

#Usuń wpis
del super_bohaterowie['Wizard']
print(super_bohaterowie)

#Zastąp wartość
super_bohaterowie['Wizard']='Spider Man'

#Wyświetl liczbę elementów w słowniku
print(len(super_bohaterowie))

#Zastąp wartość
super_bohaterowie['Wizard'] = 'Spider Man'

# Wydrukuj(Wyświetl) liczbę elementów w słowniku
print(len(super_bohaterowie))

# Uzyskaj listę kluczy słownika
print(super_bohaterowie.keys())

# Uzyskaj listę wartości słownikowych
print(super_bohaterowie.values())

#Warunki

# Instrukcje if, else i elif są używane do różnych działań - akcje oparte na warunkach
# Operatory porównania: ==,!=,>,<,>=,<=

# Instrukcja if wykona kod, jeśli warunek zostanie spełniony
# Białe znaki służą do grupowania bloków kodu w Pythonie
# Użyj tej samej liczby następnych spacji dla bloków kodu

wiek = 28
if wiek > 18:
    print('Możesz robić prawko!')


#Użyj instrukcji if - else, jeśli chcesz niezależnie wykonać inny kod

if wiek > 18:
    print('Możesz robić prawko!')
else:
    print('Jesteś za młody na zostanie kierowcą!')

#Jeśli chcesz sprawdzić wiele warunków, użyj elif
#Jeślij pierwszy jest prawdą, nie sprawdzi innych warunków

if ((wiek >=1) and (wiek <= 35)):
    print("Masz urodziny!")
elif (wiek == 35) or (wiek >= 65):
    print("Masz urodziny!")
elif not (wiek == 30):
    print("Nie masz urodzin")
else:
    print("Masz urodziny teraz!")

#Pętla FOR
#Pozwala wykonać akcję określoną liczbę razy

for x in range(0, 20):
    print(x, ' ', end="")

print('\n')

# Możesz użyć pętli for - aby przejść przez listę

lista_zakupów = ['Jajka', 'Ser', 'Pomidorki', 'Chilli']

for y in lista_zakupów:
    print(y)


# Możesz także zdefinować listę numerów do przejścia przez wszystkie elementy
lista_liczb = [[2,5,10],[20,50,100],[200,500,1000]];

for x in range(0, 3):
    for y in range(0,3):
        print(lista_liczb[x][y])

#Pętla WHILE
#Użyteczna gdy nie wiesz z wyprzedzeniem, ile razy będziesz musiał/a wykonać pętlę

losowa_liczba = random.randrange(0, 1000)

while (losowa_liczba != 500):
    print(losowa_liczba)
    losowa_liczba = random.randrange(0,1000)


#Interator pętli while jest zdefiniowany przed pętlą!

i=0
while (i <= 10):
    if (i%2==0):
        print(i)
    elif (i == 9):
        #Wymusza zakończenie wszystkich pętli
        break
    else:
        #Skrót dla i=i+1 ->
        i += 1
        #Przechodzi do następnej iteracji pętli
        continue

    i += 1

#Funkcje
#Funkcje pozwalają na ponowne użycie i napisanie czytelnego kodu
#Wpisz def (zdefiniuj) nazwę funkcji i otrzymane parametry
#return służy do zwrócenia parametru lub wartości

def dodajLiczby(liczba1, liczba2):
    suma = liczba1 + liczba2
    return suma

print(dodajLiczby(5,10))

#Jeśli zdefiniujesz zmienną poza funkcją, będzie ona działać w każdym miejscu
suma = 0

def roznicaLiczb(liczba1, liczba2):
    roznica = liczba1 - liczba2
    return roznica

print(roznicaLiczb(21,6))


#I/O Wejście
print('Podasz mi swoje imię?')

#Przechowuje wszystko, co wpisano aż do wywołania przez -> ENTER
imie = sys.stdin.readline()

print('Cześć, ', imie)

#Stringi (Ciągi znaków)
dlugi_string = "martw się, co inni ludzie o Tobie myślą, a zawsze będziesz ich więzniem. - Lao Tzu"

#Pobierz pierwsze 4 znaki sentencji
print(dlugi_string[0:4])

#Wyświetl ostatnie 5 znaków
print(dlugi_string[-5:])

#Wszystko do ostatnich 5 znaków
print(dlugi_string[:-5])

#łaczy części stringa
print(dlugi_string[:4] + " <-> epicki cytat")

#Formatowanie Stringów
print("%c to moja litera% s, a numer% d to% .5f"% ('X', 'ulubiona', 1, .14))

#Pierwsza litera zaczyna się wielką literą
print(dlugi_string.capitalize())

#Zwraca indeks początku łańcucha
#rozróżniana wielkość liter

print(dlugi_string.find("ludzie"))

#Zwraca prawdę, jeśli wszystkie znaki są literami | -> nie jest literą
print(dlugi_string.isalpha())

#Zwraca długość Stringa
print(len(dlugi_string))

#Zastąp pierwsze słowo drugim (Dodaj liczbę, aby zastąpić więcej)
print(dlugi_string.replace("ludzie","   konkurenci   "))

#Usuń spacje z przodu i końca
print((dlugi_string.strip()))

#Podziel ciąg na listę na podstawie podanego separatora
cytat_lista = dlugi_string.split(" ")
print(cytat_lista)



#Pliki I/O

#Napisz lub utwórz plik do zapisu
testowy_plik = open("test.text", "wb")

#Pobierz używany tryb pliku
print(testowy_plik.mode)

#Wpisz tekst do pliku z nową linią
testowy_plik.write(bytes("Jak zajac programiste?\n- Przeczytaj zdanie ponizej.\n- Przeczytaj zdanie powyzej.\n","UTF-8"))

#Zamyka plik
testowy_plik.close()

#Otwiera plik do odczytu i zapisu
testowy_plik = open("test.text", "r+")

#Przeczytaj/ Wyświetl tekst z pliku
testowy_plik = testowy_plik.read()

print(testowy_plik)

print(testowy_plik.find("zdanie"))

#CLASSES AND OBJECTS / Klasy i objekty
#Koncepcja Programowania Obiektowego (OOP) pozwala nam modelować rzeczy ze świata rzeczywistego za pomocą kodu
#Każdy obiekt ma pewne atrybuty (np. kolor, wysokość, waga), które są zmiennymi obiektu

"""
Klasy umożliwiają łaczenie danych i funkcji. Utworzenie nowej klasy tworzy nowy typ obiektu,
umozliwjając tworzenie nowych instancji tego typu.
Do każdej instancji klasy mogą być dołączone atrybuty służace do utrzymywania jej stanu.
Instancje klasy mogą rownież mieć metody (zdefiniowane przez jej klasę) do modyfikowania stanu.
"""

#Obiekt jest po prostu zbiorem danych (zmiennych) i metod (funkcji), które działają na tych danych.
#Podobnie klasa jest planem dla tego obiektu

class Czlowiek:

#"None" - oznacza brak wartości
#Możesz ustawić zmienną jako prywatną, zaczynając od __

    __imie = None
    __wzrost = None
    __waga = None
    __jezyk = None

#Konstruktor jest wywoływany w celu ustawienia lub zainicjowania obiektu

#self -> pozwala obiektowi odwoływać się do siebie węwnątrz klasy

    def __init__(self, imie, wzrost, waga, jezyk):
        self.__imie = imie
        self.__wzrost = wzrost
        self.__waga = waga
        self.__jezyk = jezyk


    def ustaw_imie(self, imie):
        self.__imie = imie

    def ustaw_wzrost(self, wzrost):
        self.__wzrost = wzrost

    def ustaw_waga(self, waga):
        self.__waga = waga

    def ustaw_jezyk(self, jezyk):
        self.__jezyk = jezyk

    def pobierz_imie(self):
        return self.__imie

    def pobierz_wzrost(self):
        return self.__wzrost

    def pobierz_waga(self):
        return self.__waga

    def pobierz_jezyk(self):
        return self.__jezyk

    def pobierz_typ(self):
        print("Człowiek")

    def naString(self):
        return "{} ma {} cm wzrostu i waży {}kg, a programuje w {}".format(self.__imie, self.__wzrost, self.__waga, self.__jezyk)

#Jak stworzyć czlowieka?

programista = Czlowiek('Kris', 175, 70,'C#, Java, Scala, Python')

print(programista.naString())

#Nie możesz uzyskać bezpośredniego dostępu do tej wartości, ponieważ jest ona prywatna ->
#print(programista.__imie)
#print(programista.imie)



#Inheritance / Dziedziczenie

#Możesz dziedziczyć wszystkie zmienne i metody z innej klasy - przykład poniżej:

class Kierowca(Czlowiek):
    __firma = None

    def __init__(self, imie, wzrost, waga, jezyk, firma):
        self.__firma = firma
        self.__pobierz_typ = None

        #Jak wywołać konstruktora superklasy?
        super(Kierowca, self).__init__(imie,wzrost,waga,jezyk)


    def ustaw_firma(self, firma):
         self.__firma = firma

    def pobierz_firma(self):
        return self.__firma

    def pobierz_typ(self):
        print("Kierowca")


    #W ten sposób możemy nadpisywać funkcje w superklasie ->

    def naString(self):
         return "{} ma {} cm wzrostu i waży {}kg, a programuje w {}. Pracuje w firmie {}".format(self.pobierz_imie(),
                                                                                                 self.pobierz_wzrost(),
                                                                                                 self.pobierz_waga(),
                                                                                                 self.pobierz_jezyk(),
                                                                                                 self.__firma)


mario = Kierowca("Mariusz",190, 96, "- troche ogarnia pythona", "DHL")

print(mario.naString())

#Polimofizm pozwala na określenie obiektów jako ich superklasy ->

class TestCzlowieka:
    def pobierz_typ(self,czlowiek):
        czlowiek.pobierz_typ()

testowany_czlowiek = TestCzlowieka()

testowany_czlowiek.pobierz_typ(programista)

testowany_czlowiek.pobierz_typ(mario)
```
