```python
# 讓我們做一個戰爭遊戲
# 英雄 --> 名稱 - 攻擊力 - 生命值
# 英雄 --> 攻擊/防禦
# 想要攻擊的英雄將在[0-30]之間選擇他想要攻擊的力量。 高命中將具有低攻擊機會。
# 如果對方被攻擊無法反應，則對方的HP會因攻擊量而減少
# 如果對手可以反擊，英雄的攻擊將減少生命值
# 如果你的對手的生命值為0，他將贏得戰鬥
```

```python
import random

class Hero:
    def __init__(self, name):
        self.Name = name
        self.HealthPoint = 100
        
    def attack(self, hero, power, turn):
        if power < 10:
            hit_ratio = 0.9
        elif 10 <= power < 15:
            hit_ratio = 0.8
        elif 15 <= power < 25:
            hit_ratio = 0.5
        else:
            hit_ratio = 0.3
        if random.random() < hit_ratio:
            print(25*" ", self.Name, hero.Name, "'yeah!", power)
            self.hit(hero, power) 
        else:
            print(30*" ", self.Name, "missed!", power)
            self.miss(power)
        self.print_info(hero, turn)
    
    def hit(self, hero, power):
        hero.HealthPoint -= power
        
    def miss(self, power):
        self.HealthPoint -= power
        
    def print_info(self, hero, turn):
        hero1 = self
        hero2 = hero
        if turn == 1:
            hero1 = self
            hero2 = hero
        elif turn == 2:
            hero1 = hero
            hero2 = self
        
        print(hero1.Name, " "*(77-len(hero1.Name) - len(hero2.Name)), hero2.Name)
        print(hero1.HealthPoint//3 * "|", " "*(77-hero1.HealthPoint//3 - hero2.HealthPoint//3), hero2.HealthPoint//3 * "|")


name1 = input("Please enter the name of the 1st Hero: ")
name2 = input("Please enter the name of the 2nd Hero: ")

hero1 = Hero(name1)
hero2 = Hero(name2)

print(30*" ", "THE GAME BEGINS")
turn = 1

while hero1.HealthPoint > 0 and hero2.HealthPoint > 0:
    if turn == 1:
        power = input(hero1.Name + " how much power does she want to hit?")
        hero1.attack(hero2, int(power), turn)
        turn = 2
        
    elif turn == 2:
        pwrs = [5, 10, 15, 20, 25, 30]
        idx = random.randint(0, len(pwrs)-1)
        hero2.attack(hero1, pwrs[idx], turn)
        turn = 1
        
if hero1.HealthPoint <= 0:
    print(30*" ", hero2.Name, "WON")
else:
    print(30*" ", hero1.Name, "LOSE")
```
