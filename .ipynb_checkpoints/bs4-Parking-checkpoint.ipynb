```python
f = open("input/json-桃園公共自行車即時服務資料.json", "r", encoding='UTF-8')

x = json.loads(f.read())
data = x['retVal']
key = data.keys()

for k in key:
    d = data[k]
    print(d["sna"],d["ar"],d["sbi"],"/",d["tot"])
```
