```python
import pandas as pd

df = pd.read_csv('input/SalesRecords.csv')
df[0:5]
```

```python
del(df['Total Revenue'],df['Total Cost'])
df
```

```python
df.tail(5)
```

```python
df['Total Revenue']=df['Units Sold']*df['Unit Price']
df
```

```python
df['Total Revenue']=df['Units Sold']*df['Order Priority']
df
```

```python
df['Length']=len(df['Total Revenue'])
df
```

Apply Lambda

```python
df['Length']=df['Total Revenue'].apply(lambda x:len(x))
df
```

```python
df['Year']=df['Order Date'].apply(lambda x:x.split('/')[2])
df
```

```python
df['High Cost']=df['Unit Cost'].apply(lambda x:True if x>100 else False)
df
```

```python
group=df.groupby(['Country','Sales Channel'])
group['Units Sold'].agg({sum})
```
