```python
import numpy as np
import pandas as pd

data = pd.DataFrame({'Name':['A','A','A','B','B','B','C','C','C'],
                    'F1':[11,12,13,14,15,16,17,18,19],
                    'F2':[21,22,23,24,25,26,27,28,29],})
data
```

## Aggregation 

```python
def AA(x):
    global i
    print(x)
    print(str(i)+'==================\n')
    i = i+1
    
i = 1
data.groupby('Name')['F1'].aggregate(AA)
```

```python
def AA(x):
    global i
    print(x)
    print(x.mean())
    print(str(i)+'==================\n')
    i = i+1
    return x.mean()
    
i = 1
data.groupby('Name')['F1'].aggregate(AA)
```

```python
def AA(x):
    global i
    print(x)
    print(x.mean())
    print(str(i)+'==================\n')
    i = i+1
    return x.mean()
    
i = 1
data.groupby('Name')[['F1','F2']].aggregate(AA)
```

## Transform

```python
def AA(x):
    global i
    print(x)
    print(x.count())
    print(str(i)+'==================\n')
    i = i+1
    return x.count()
    
i = 1
data.groupby('Name')['F1'].transform(AA)
```

## apply

```python
def AA(x):
    global i
    print(x)
    print(x.mean())
    print(str(i)+'==================\n')
    i = i+1
    return x
    
i = 1
data.groupby('Name')[['F1','F2']].apply(AA)
```

```python
def AA(x):
    global i
    x['F3'] = x['F1'].mean() * x['F2'].mean()
    print(x)
    print(str(i)+'==================\n')
    i = i+1
    return x
    
i = 1
data.groupby('Name')[['F1','F2']].apply(AA)
```

## filter

```python
def AA(x):
    global i
    print(x)
    print(str(i)+'==================\n')
    i = i+1
    return i%2 == 0
    
i = 1
data.groupby('Name')[['F1','F2']].filter(AA)
```

```python
def AA(x):
    global i
    print(x)
    print(np.mean(x))
    print(str(i)+'==================\n')
    i = i+1
    return np.mean(x)['F1']>15
    
i = 1
data.groupby('Name')[['F1','F2']].filter(AA)
```

```python
def AA(x):
    global i
    print(x)
    print(np.mean(x))
    print(str(i)+'==================\n')
    i = i+1
    return np.mean(x)
    
i = 1
d = data.groupby('Name')[['F1']].transform(AA)
d
```
