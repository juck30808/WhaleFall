```python
# 幣別的平均匯率與最低跟最高匯率

f = open("input/json-國際主要國家貨幣每月匯率概況.json", "r", encoding='UTF-8')
data = json.loads(f.read())

def exchange(country):
    print(country +'匯率')
    sum = 0
    for d in data:
        v = float(d[country])
        if v > 0:
            max = v
        if v < 9999999:
            min = v
        sum += v
    print('平均',sum / len(data))
    print('最高', max)
    print('最低', min)
    print('')


exchange('新台幣')
exchange('人民幣')
```
