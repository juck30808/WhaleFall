## numpy.array

```python
import numpy as np
a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6]])
```

```python
# sort
np.concatenate((a, b), axis=0)
# array([[1, 2],
#        [3, 4],
#        [5, 6]])
```

```python
# sort by axis 1
np.concatenate((a, b.T), axis=1) # b有轉置
# array([[1, 2, 5],
#        [3, 4, 6]])
```

```python
np.concatenate((a, b), axis=None)
# array([1, 2, 3, 4, 5, 6])
```

# 2.Data filter

```python
list1 = [1,2,3,4,5,6]
list2 = [x for x in list1 if x % 2 == 1] # 奇數
list2 
# [1, 3, 5]
```

```python
list1 = [1,2,3,4,5,6]
list2 = list(filter(lambda x : x % 2 == 1, list1))
list2 
# [1, 3, 5]
```

```python
dict1 = {"a":1, "b":2, "c":3}

#filter a and c
dict2 = {k:v for k,v in dict1.items() if k in ["a","c"]}
dict2

# filter value
dict3 = {k:v for k,v in dict1.items() if v%2 == 1} 
dict3
```

```python
import pandas as pd
df = pd.DataFrame({"a":[1,2,3], 
                   "b":[4,5,6], 
                   "c":[7,8,9], 
                   "d":[True, True,False]})
df
```

```python
# filter a,c col
df[["a","c"]]  #filter a & c
```

```python
df[df["d"]]    #filter d is True
```

# 3.data sort

```python
A = [1,3,5,1,2]
sorted(A)
```

```python
A = [1,3,5,1,2]
sorted(A,reverse=True)
```

```python
B = [("Alice",100), ("Bob", 97), ("Carol", 97), ("Bob", 95) ]
sorted(B)
```

```python
B = [("Alice",100), ("Bob", 97), ("Carol", 97), ("Bob", 95) ]
sorted(B,key=lambda x: x[1]) # sort by x[1]
```

```python
C = ["Alice", "Bob", "Bob", "Bob", "Carol"]
sorted(C,key=len)
```
