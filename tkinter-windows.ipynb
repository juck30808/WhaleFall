```python
# 加入按鈕元件的反應
import tkinter as tk
def onButtonClick():
    label['text'] = 'OH!'
    

window = tk.Tk()
window.title("這是一個練習用的程式")
window.geometry("300x200+250+150") # 寬x高+起始位置x軸+起始位置y軸

# -------------
label = tk.Label(window, 
                 text = 'Hello!', # 顯示的文字
                 bg = '#CCCC00',  # 背景顏色
                 font = ('Arial', 24), # 顯示字型跟大小
                 width = 10, height = 2 # 元件大小
                ) #註冊一個 LABEL 元件在視窗上
label.pack() #將LABEL元件顯示出來

button = tk.Button(window, text = '按我一下!',
                   command = onButtonClick #按下按鈕時的反應
                  )
button.pack()
window.mainloop()
```

```python
# 如果有兩個以上的按鈕時
import tkinter as tk


def onButton1Click():
    global fontSize #將這個變數宣告成全域型
    fontSize += 2
    label['font'] = ('Arial', fontSize)
def onButton2Click():
    global fontSize #將這個變數宣告成全域型
    fontSize -= 2
    label['font'] = ('Arial', fontSize)
    

fontSize = 24
window = tk.Tk()
window.title("這是一個練習用的程式")
window.geometry("300x200+250+150") # 寬x高+起始位置x軸+起始位置y軸
label = tk.Label(window, 
                 text = 'Hello!', # 顯示的文字
                 bg = '#CCCC00',  # 背景顏色
                 font = ('Arial', fontSize), # 顯示字型跟大小
                 width = 10, height = 2 # 元件大小
                ) #註冊一個 LABEL 元件在視窗上
label.pack() #將LABEL元件顯示出來

button1 = tk.Button(window, text = '變大',
                   command = onButton1Click #按下按鈕時的反應
                  )
button2 = tk.Button(window, text = '變小',
                   command = onButton2Click #按下按鈕時的反應
                  )
button1.pack()
button2.pack()
window.mainloop()
```

```python
# 如果有兩個以上的按鈕時
import tkinter as tk


def onButtonClick(unit):
    global fontSize #將這個變數宣告成全域型
    fontSize += unit
    label['font'] = ('Arial', fontSize)

fontSize = 24
window = tk.Tk()
window.title("這是一個練習用的程式")
window.geometry("300x200+250+150") # 寬x高+起始位置x軸+起始位置y軸
label = tk.Label(window, 
                 text = 'Hello!', # 顯示的文字
                 bg = '#CCCC00',  # 背景顏色
                 font = ('Arial', fontSize), # 顯示字型跟大小
                 width = 10, height = 2 # 元件大小
                ) #註冊一個 LABEL 元件在視窗上
label.pack() #將LABEL元件顯示出來

button1 = tk.Button(window, text = '變大',
                   command = onButtonClick(+2) #按下按鈕時的反應
                  )
button2 = tk.Button(window, text = '變小',
                   command = onButtonClick(-2) #按下按鈕時的反應
                  )
button1.pack()
button2.pack()
window.mainloop()
```

```python
''' 
修改元件的位置
以下程式參考自:
https://blog.techbridge.cc/2019/09/21/how-to-use-python-tkinter-to-make-gui-app-tutorial/
'''
import tkinter as tk
import math

window = tk.Tk()
window.title('BMI App')
window.geometry('600x200')
window.configure(background='white')

def calculate_bmi_number():
    height = float(height_entry.get())
    weight = float(weight_entry.get())
    bmi_value = round(weight / math.pow(height, 2), 2)
    result = '你的 BMI 指數為：{} {}'.format(bmi_value, get_bmi_status_description(bmi_value))
    result_label.configure(text=result)

def get_bmi_status_description(bmi_value):
    if bmi_value < 18.5:
        return '體重過輕囉，多吃點！'
    elif bmi_value >= 18.5 and bmi_value < 24:
        return '體重剛剛好，繼續保持！'
    elif bmi_value >= 24 :
        return '體重有點過重囉，少吃多運動！'

header_label = tk.Label(window, text='BMI 計算器')
header_label.pack()

height_frame = tk.Frame(window)
height_frame.pack(side=tk.TOP)
height_label = tk.Label(height_frame, text='身高（m）')
height_label.pack(side=tk.LEFT)
height_entry = tk.Entry(height_frame)
height_entry.pack(side=tk.LEFT)

weight_frame = tk.Frame(window)
weight_frame.pack(side=tk.TOP)
weight_label = tk.Label(weight_frame, text='體重（kg）')
weight_label.pack(side=tk.LEFT)
weight_entry = tk.Entry(weight_frame)
weight_entry.pack(side=tk.LEFT)

result_label = tk.Label(window)
result_label.pack()

calculate_btn = tk.Button(window, text='馬上計算', command=calculate_bmi_number)
calculate_btn.pack()

window.mainloop()
```
